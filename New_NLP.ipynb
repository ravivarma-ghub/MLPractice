```python
#Stemming
'''Stemming is the process of reducing a word to its word stem, which may involve removing affixes, suffixes, or prefixes,
 or reducing it to the root of the word known as the lemma. This definition is sourced from Wikipedia.
 Stemming plays a crucial role in natural language understanding and processing.
'''
```

```python
from nltk.stem import PorterStemmer

porter = PorterStemmer()

words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized"]

for word in words:
    print(f"{word} -> {porter.stem(word)}")
```

```python
#stemming might change meaning works for very good number of words but fails for some
#For some words the meaning might change after stemming
#We have other Stemming algorithms as well like RegexpStemmer, SnowballStemmer etc.
```

```python
from nltk.stem import RegexpStemmer
reg=RegexpStemmer('ing$|s$|ed$|$able',min=4)
reg.stem('eating')
```

```python
reg.stem('ingeating')
```

```python
from nltk.stem import SnowballStemmer
snowball=SnowballStemmer('english')
for word in words:
    print(f"{word} -> {snowball.stem(word)}")
```

WordNet Lemmatizer
Lemmatization is a technique similar to stemming, but it aims to produce the exact root word, called a lemma, which is a valid word

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running","v"))
```
