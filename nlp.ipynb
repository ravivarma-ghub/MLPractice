```python
'''Stemming
Stemming is the process of reducing a word to its word stem, which may involve removing affixes, suffixes,
 or prefixes to reach the root of the word, known as the lemma. This process is crucial in natural language understanding and processing.

For example, words like "eating," "eat," and "eaten" share the root word "eat." Similarly, "going," "gone," and "goes" share the root "go."
 Stemming helps consolidate these variations into a single form, reducing the number of unique input features in text data.'''
```

```python
##Porter Stemming
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
words = ['eating', 'eats', 'eaten', 'writing', 'writes', 'programming', 'programs', 'history', 'finally', 'finalized']

for word in words:
    print(word + " -> " + stemmer.stem(word))
```

```python
import pandas as pd
```

```python
data = pd.read_csv('spam.csv', encoding='latin-1')
```

```python
data.head()
data.columns
```

```python
data.drop([ 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True)
```

```python
data.head()
```

```python
#Data cleaning and preprocessing 
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
import re
```

```python
ps=PorterStemmer()
```

```python
corpus = []
for i in range(0, len(data)):
    review = re.sub('[^a-zA-Z]', ' ', data['v2'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#Create the bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
```

```python
x=cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
cv.vocabulary_
```

```python
#N-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=250,ngram_range=(2,2))
x=cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
cv.vocabulary_
```

```python
#Create TF-IDF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf=TfidfVectorizer(max_features=250)
xx=tf.fit_transform(corpus).toarray()
```
