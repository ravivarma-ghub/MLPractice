```python
df_copy=df.copy()
```

```python
df_copy.drop(index=10472,inplace=True)
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews']=pd.to_numeric(df_copy['Reviews'],errors='coerce')
```

```python
df_copy['Size']=df_copy['Size'].replace('M','000')
df_copy['Size']=df_copy['Size'].replace('K','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=pd.to_numeric(df_copy['Size'],errors='coerce')
```

```python

df['Installs'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for col in cols_to_clean:
    for char in chars_to_remove:
        df_copy[col]=df_copy[col].replace(char,'')
```

```python
df_copy['Installs'] = pd.to_numeric(df_copy['Installs'], errors='coerce')
df_copy['Price'] = pd.to_numeric(df_copy['Price'], errors='coerce')
```

```python
df_copy['Reviews']=df_copy['Reviews'].astype(int)
df_copy['Installs']=df_copy['Installs'].astype(float)
```

```python
df_copy.info()
```

```python
df['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['month']=df_copy['Last Updated'].dt.month
df_copy['year']=df_copy['Last Updated'].dt.year
df_copy['day']=df_copy['Last Updated'].dt.day
df_copy.info()
```

```python
df_copy.to_csv('GoogleCleaned.csv')
```
